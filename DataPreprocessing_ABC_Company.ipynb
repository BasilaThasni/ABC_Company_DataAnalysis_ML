#4. Discover which team and position have the highest salary expenditure.

```python
# Find the salary expenditure of each team
highest_salary_expenditure_team_df = salary_by_team.groupby(['Team'], as_index=False)['Salary'].sum() # group by team and sum salary of each team
highest_salary_expenditure_team_df.head()
```

```python
highest_salary_expenditure_team_df.loc[highest_salary_expenditure_team_df['Salary'].idxmax()] #Shows the team with the highest salary expenditure with the total salary amount.
```

```python
high_salary_team_data = highest_salary_expenditure_team_df.loc[highest_salary_expenditure_team_df['Salary'].idxmax()]
high_salary_team_data.iloc[0]
```

```python

print(f"Team having the highest salary expenditure is {high_salary_team_data.iloc[0]}")
print(f"The salary expenditure is ${high_salary_team_data.iloc[1]}")
```

```python
salary_by_team[salary_by_team['Team'] == 'Cleveland Cavaliers']
```

```python
plt.figure(figsize=(12, 8))
plt.bar(highest_salary_expenditure_team_df.Team, highest_salary_expenditure_team_df.Salary, color='blue')
plt.title('Team Salary Expenditure', fontsize=14, fontweight='bold')
plt.xlabel('Team', fontsize=14)
plt.ylabel('Salary Expenditure', fontsize=14)
plt.xticks(rotation=90)
highest_salary = highest_salary_expenditure_team_df.Salary.max()
highest_salary_index = highest_salary_expenditure_team_df.Salary.idxmax()
plt.text(highest_salary_index, highest_salary/2, '$'+str(highest_salary), color='white', ha='center', va='bottom',rotation=90)
plt.tight_layout()
plt.show()
```

#5. Investigate if there's any correlation between age and salary, and represent it visually.

```python
correlation = df[['Age', 'Salary']].corr() # calculate correlation metrix
correlation # show
```

```python
# Scatter plot of Age vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Team', palette="viridis")
plt.title("Correlation between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='viridis')
plt.title('Correlation between Age and Salary', fontsize=14, fontweight='bold')
plt.show()
```

#The heatmap shows the correlation between age and salary of employees of ABC Company.

##Low Positive Correlation: The correlation coefficient between Age and Salary is 0.21. This indicates a weak positive correlation, meaning that as the age of employees slightly increases, their salary also tends to increase, but not very strongly.

##Interpretation of Value: A correlation value of 0.21 suggests that age does not have a significant impact on the salary of the employees.

##The rate of salary might also depend upon the experience and skills each employee has.

#Hence, age and salary of the employees have a minimal positive association, implying that age is not a strong determinant of salary in this context.
